```python
import matplotlib.pyplot as plt
import numpy as np
# import matplotlib.patches as patches

X = list_coord = np.random.rand(10000, 2)

figure, axes = plt.subplots()
draw_circle = plt.Circle((0, 0), 1, fill = False)
axes.add_artist(draw_circle)

A = 0
B = 0

for i in range(10000):
    Dot = list_coord[i]
    
    if Dot[0]**2 + Dot[1]**2 <= 1:
        A += 1
        plt.scatter(Dot[0], Dot[1], s= 1, color= 'blue')
    else:
        B += 1
        plt.scatter(Dot[0], Dot[1], s= 1, color= 'red')

print(A)
print(B)
print(A+B)
print(A/(A+B)*4)

plt.title("Monte Carlo - Pi Calculation")
plt.xlim(0, 1) and plt.ylim(0, 1)
plt.show()
```
